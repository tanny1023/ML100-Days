```python
import os
os.environ['PROJ_LIB'] = 'D:\anaconda3\pkgs\proj-7.2.0-h3e70539_1\Library\share'
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(8, 6), edgecolor='w')
m = Basemap(projection = 'ortho', lat_0 = 0, lon_0 = 0)
m.bluemarble()
plt.show()
```

```python
m = Basemap(lat_0=50, lon_0=-107, projection='lcc', resolution=None, width=12000000, height=9000000, lat_1=45, lat_2=55)
m.bluemarble()
plt.show()
```

```python
m = Basemap(projection='mill', llcrnrlat=-90, llcrnrlon=-180, urcrnrlat=90, urcrnrlon=180)
m.drawcoastlines()
m.drawcountries(linewidth=2)
m.drawstates(color='b')
m.drawcounties(color='darkred')
plt.title('Basemap Tutorial')
plt.show()
```
