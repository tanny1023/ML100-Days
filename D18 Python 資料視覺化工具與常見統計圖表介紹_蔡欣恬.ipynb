# 範例重點
如何使用亂數, 資料集來操作

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
```

```python
plt.plot(x,y)
plt.show() 
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
```

```python
plt.plot(x, y_sin)
plt.show()
```
