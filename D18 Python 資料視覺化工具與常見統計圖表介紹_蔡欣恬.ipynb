# 範例重點
如何使用亂數, 資料集來操作

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
```

```python
#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
plt.plot(x, y_sin)
plt.show()
```

```python
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
plt.plot(x, y_cos)
plt.show()
```

```python
#劃出散點圖: Scatter Plots，顏色由（X，Y）的角度給出。注意標記的大小，顏色和透明度。
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X)
plt.scatter(X,Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
