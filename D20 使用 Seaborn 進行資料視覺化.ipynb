# Seaborn圖樣式

```python
import numpy as np
import seaborn as sns
```

```python
#設定圖形樣式 - whitegrid
# use sns.set
sns.set(style="whitegrid")
```

```python
rs = np.random.RandomState(7)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)
```

```python
sns.residplot(x, y, lowess=True, color="g")
```
